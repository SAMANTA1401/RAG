```python

# You're iterating over the chat history and displaying each message using Markdown formatting.
# Chat History Iteration:
# The for loop iterates over the chat.history attribute, which contains a list of messages exchanged in the chat.
# Message Attributes:
# Each message object has the following attributes:
# role: The role of the message sender (e.g., user or AI).
# parts: A list of message parts, where each part contains the message text.
# Markdown Display:
# The display function is used to display the message text using Markdown formatting. The Markdown string is created using an f-string, which inserts the message.role and message.parts[0].text values into the string.
```

```python
for message in chat.history:
    display(Markdown(f'**{message.role}**: {message.parts[0].text}'))
```
