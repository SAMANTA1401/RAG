```python
url = 'https://medium.com/enterprise-rag/a-first-intro-to-complex-rag-retrieval-augmented-generation-a8624d70090f'
```

```python
document = SimpleWebPageReader(html_to_text=True).load_data(urls=[url])
# document = SimpleDirectoryReader(input_dir='data').load_data()
```

```python
index = VectorStoreIndex.from_documents(documents=document)
```
